```python
import pandas as pd
import numpy as np
df = pd.read_excel('data.xlsx')
df.describe()

```

```python


bins = np.arange(116.5, 255.5, 13)
print(bins)
print()
df['group'] = pd.cut(df['data'],bins,right=False)
counts =df['group'].value_counts(sort=True)
per = df['group'].value_counts(normalize=True,sort=True)
print(counts)
print()
print(per)

scoregroup = df.groupby('group')

list(scoregroup)
```
